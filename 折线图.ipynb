```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示为方块的问题

# 年度销售数据
months = ['一月', '二月', '三月', '四月', '五月', '六月', '七月', '八月', '九月', '十月', '十一月', '十二月']
sales = [10000, 12000, 15000, 11000, 13000, 14000, 16000, 18000, 17000, 19000, 20000, 21000]

# 创建折线图
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', color='skyblue', linestyle='-', linewidth=2)

# 在每个数据点处显示具体的销售额
for i, sale in enumerate(sales):
    plt.text(i, sale, sale, ha='center', va='bottom')

# 添加标题和标签
plt.title('年度销售报告')
plt.xlabel('月份')
plt.ylabel('销售额（单位：元）')

# 显示折线图
plt.show()
```
