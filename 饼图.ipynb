```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 设置全局中文字体为Microsoft YaHei
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']

# 年度销售数据
months = ['一月', '二月', '三月', '四月', '五月', '六月', '七月', '八月', '九月', '十月', '十一月', '十二月']
sales = [10000, 12000, 15000, 11000, 13000, 14000, 16000, 18000, 17000, 19000, 20000, 21000]

# 创建饼图
plt.figure(figsize=(8, 8))
patches, texts, autotexts = plt.pie(sales, labels=months, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# 在饼图外显示销售额标识
plt.legend(patches, [f'{month}: {sale}元' for month, sale in zip(months, sales)], loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# 添加标题
plt.title('年度销售情况')

# 显示饼图
plt.show()
```
