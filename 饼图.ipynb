```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# Microsoft YaHei
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']

# 
months = ['', '', '', '', '', '', '', '', '', '', '', '']
sales = [10000, 12000, 15000, 11000, 13000, 14000, 16000, 18000, 17000, 19000, 20000, 21000]

# 
plt.figure(figsize=(8, 8))
patches, texts, autotexts = plt.pie(sales, labels=months, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# 
plt.legend(patches, [f'{month}: {sale}' for month, sale in zip(months, sales)], loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# 
plt.title('')

# 
plt.show()
```
